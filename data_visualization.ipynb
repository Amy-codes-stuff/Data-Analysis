```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1,2,3,4]
y = [10,20,30,40]
plt.figure(figsize = (5,4))
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Plot')
plt.plot(x,y, color = 'c', linewidth = 3, marker = 'o', markerfacecolor = 'red', alpha = 1)   #alpha  = transparancy
plt.grid(color = 'blue')
plt.show()
```

```python
plt.style.use('dark_background')
x = [1,2,3,4]
y = [10,20,30,40]
plt.figure(figsize = (5,4))
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Plot')
plt.plot(x,y, color = 'c', linewidth = 3, linestyle = '--' ,marker = 'o', markerfacecolor = 'red', alpha = 1)   #alpha  = transparancy
plt.grid(color = 'blue')
plt.show()
```

```python
#bar plot

x = ['A','B','C','D']
y = [20,40,60,80]
plt.figure(figsize = (5,4))
plt.grid(False)
plt.bar(x,y, color = 'c', edgecolor = 'green', linewidth = 3, width = 0.3  )
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Bar Plot')
plt.show()
```

```python
#bar plot horizontal

x = ['A','B','C','D']
y = [20,40,60,80]
plt.figure(figsize = (5,4))
plt.grid(False)
plt.barh(x,y, color = 'c', edgecolor = 'green', linewidth = 3, height = 0.5  )
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Bar Plot')
plt.show()
```

```python
# scatter plot

x = np.random.randint(1, 50, 20)
y = np.random.randint(1, 50, 20)
plt.figure(figsize = (5,4))

plt.scatter(x,y)
plt.show()
```

```python
#histogram
plt.figure(figsize = (5,4))
x = np.random.rand(50)
plt.hist(x, bins = 10)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('histogram')
plt.show()
```

```python
#subplots
plt.figure(figsize = (5,4))

x = [1,2,3,4]
y = [10,20,30,40]
z = [5,10,15,20]

#first subplot
plt.subplot(1,2,1)
plt.plot(x,y , color = 'c')
plt.title('First Plot')

#second subplot
plt.subplot(1,2,2)
plt.plot(x,z, color = 'r')
plt.title('Second Plot')

plt.show()
```

```python
# heatmap

x = np.random.rand(4,4)
plt.imshow(x, cmap = 'Spectral', interpolation='bicubic')
plt.colorbar()
plt.show()
```

```python
df = pd.read_csv('sales_data.csv')
```

```python
df.head()
```

```python
#total revenue based on product category
filter_data = df.groupby('Product Category')['Total Revenue'].sum()
plt.bar(filter_data.index, filter_data.values)
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.title('Total Revenue by Product Category')
plt.show()
```

```python
#revenue over time

df['Date'] = pd.to_datetime(df['Date'])
plt.figure(figsize = (10,6))
plt.plot(df['Date'], df['Total Revenue'], marker = 'o')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.title('Total Revenue Over Time')
plt.show()
```

```python
#plot histogram for total revenue

plt.hist(df['Total Revenue'], bins = 20, edgecolor = 'black')
plt.xlabel('Total Revenue')
plt.ylabel('Frequency')
plt.title('Distribution of Total Revenue')
plt.show()
```

```python
#piechart

region_rev = df.groupby('Region')['Total Revenue'].sum()
plt.pie(region_rev, labels = region_rev.index, colors = ['r', 'b','c'], autopct = '%1.1f%%')
plt.title('Total Revenue by Region')
plt.show()
```

```python
payment = df.groupby('Payment Method')['Total Revenue'].sum()
plt.pie(payment, labels = payment.index, colors = ['r', 'b','c'], autopct = '%1.1f%%')
plt.title('Total Revenue by Payment Method')
plt.show()
```

```python
import seaborn as sns
```

```python
sns.histplot(df['Total Revenue'], bins = 20)
plt.show()
```

```python
#boxplot
sns.boxplot(x = 'Product Category', y = 'Total Revenue', data = df)
plt.show()
```

```python
from numpy import mean
sns.barplot(x = 'Region', y = 'Total Revenue', data = df, estimator = mean)
plt.show()
```

```python
#plot for units sold for payment method
sns.barplot(x = 'Payment Method', y = 'Units Sold', data = df)
plt.show()
```

```python
sns.countplot(x = 'Payment Method', data = df)
```

```python
#top 10 products based on revenue

data = df.groupby('Product Name')['Total Revenue'].sum().sort_values(ascending = False).head(10)
sns.barplot(x = data.values, y = data.index)
plt.show()
```

```python
#average unit price by product category

fdata = df.groupby('Product Category')['Unit Price'].mean().reset_index()
sns.barplot(x = 'Product Category', y = 'Unit Price', data = fdata)
plt.show()
```
